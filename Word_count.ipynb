```python
import nltk
import docx2txt
import string
from stop_words import get_stop_words
```

```python
# Process the DOCX file
text = docx2txt.process(r"C:\Users\NTres\OneDrive - Danmarks Tekniske Universitet\Formelle_papirer\Ansoegning_Netcompany_consulting\Nikolai_ansoegning.docx")
language = "danish"

# Define stop words:
stop_words_dict = {"danish": get_stop_words('danish'), "english": get_stop_words('english')}
stop_words = stop_words_dict[language]

# Initialize the dictionary to store word counts
word_dict = dict()

# Iterate over each line of the text
for line in text.split("\n"):  # Use split("\n") to handle line breaks more effectively
    line = line.strip().lower()
    words = line.split()

    for word in words:
        # Remove punctuation from each word
        word = word.strip(string.punctuation)
        
        if word in stop_words:
            continue
        
        # Skip empty words after cleaning
        if word:
            if word in word_dict:
                word_dict[word] += 1
            else:
                word_dict[word] = 1

# Print the word counts
for key, value in word_dict.items():
    print(f"{key}: {value}")
```
